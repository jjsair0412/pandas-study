# Pandas DataFrame Study
Pandas Series는 일차원인데, ***DataFrame은 다중 차원으로 되어있음.***
- 여러 라인으로 되어있다는 의미로 볼 수 있음.

Pandas DataFrame은 Series와 다르게 한 필드에 여러값들이 들어갈 수 있음.

````python
# Pandas import
import pandas as pd
````

````python
# python dictionary로 Pandas Dataframe을 정의하기
# Pandas Dataframe 생성자 메서드로 DataFrame 생성할 수 있음
# Pandas DataFrame은 Series와 다르게 한 필드에 여러값들이 들어갈 수 있음.
bank_client_df = pd.DataFrame({'Bank Client ID':[111, 222, 333, 444],
                              'Bank Client Name':['Laila Aly','Kate Steve','Nicole Mitch' ,'Francis Morris'],
                              'Net Worth [$]':[35000, 3000, 10000, 20000],
                              'Years with Bank':[4, 7, 10, 15]})


# 4행 4열 짜리 DataFrame을 생성한것을 볼 수 있음
bank_client_df
````

````python
# DataFrame의 타입 확인하기
print(type(bank_client_df))
````

````python
# DataFrame의 처음 2행만 뽑아서 보기
bank_client_df.head(2)
````

````python
# DataFrame의 마지막에서 뒤로 몇개(1개)
bank_client_df.tail(1)
````

````python
# DataFrame의 마지막에서 뒤로 몇개(1개)
bank_client_df.tail(2)
````

````python
# DataFrame의 행 , 열 개수 출력하기
bank_client_df.shape
````

````python
# DataFrame의 메모리 사용량, null 개수, 타입 , 컬럼 등 다양한 정보를 출력해서 확인할 수 있음.
bank_client_df.info()
````

````python
portfolio_df = pd.DataFrame(data={'Stock Name':['AAPL','AMZN','MSFT'],
                   'Price':[179.66 , 178.22 , 415.50],
                   'Number Stock':[3, 5, 100]})
portfolio_df
````

````python
# 특정 열 모두 더하기, Python 내장함수와 같이 쓸 수 있음.
# . 으로 해당 열 참조 가능
sum(portfolio_df.Price)
````

````python
# head() 나 tail() 메서드를 합쳐서 사용해서 특정 행을 선택할 수 도 있음
sum(portfolio_df.Price.head(2))
````

````python
# 헤더 이름만 작성하면, 해당 열만 접근이 가능함
portfolio_df['Price']
````

````python
portfolio_df['Number Stock']
````

````python
# 같은 열 수를 가진 DataFrame이기 때문에 그냥 곱하면 곱셈이 가능함.
stock_value = portfolio_df['Price'] * portfolio_df['Number Stock']
stock_value
````

````python
# 나온 결과를 모두 더함
print('Total Portfolio value = {}'.format(stock_value.sum()))
````

# 정의한 DataFrame Export 하기
정의한 DataFrame을 output 타입 지정해서 Export 하는 방법에 대해 기술함.
- Export 할 수 있다면, execl 과 같은 툴에서 재 사용하는것이 가능함.

```.to_~()``` 메서드로 Export 할 수 있으며. html, csv, json 등 다양한 타입을 지원함.
- 파라미터로는 저장할 파일 이름을 지정함



````python
# CSV 파일을 먼저 불러온 다음 , 아웃풋으로 Export 하는 실습
bank_df = pd.read_csv('bank_client_information.csv')
bank_df
````

````python
# csv 확장자를 통해 export 해보기
# index 파라미터를 True , False 일 경우 결과값이 다름
# index를 True로 두면 , 각 행에 넘버링이 달리고(Series Index), False로 두면 안달림
bank_df.to_csv('sample_output_index.csv', index= True)
bank_df.to_csv('sample_output_index-false.csv', index= False)
````

````python
# compression 메서드로 압축 타입을 지정해서 output 되는 파일을 압축 저장할 수 있음.
bank_df.to_csv('sample_output_index_compression.gzip', index= False, compression='gzip')
````

# Pandas DataFrame Index Setting
Pandas DataFrame에서 Index를 재 설정하는 방법에 대해 기술함.


````python
# test용 csv 파일 import
# 결과값을 보면 , 숫자 Index가 기본적으로 설정되어있는것을 볼 수 있음.
# 기본적으로 DataFrame은 맨 좌측 첫번째 열로 Index가 들어감.
bank_df = pd.read_csv('bank_client_information.csv')
bank_df
````

````python
# DataFrame의 특정 열을 지정해서, 해당 열을 Index로 설정할 수 있음.
# First Name 열을 전체 DataFrame의 Index로 사용
bank_df.set_index('First Name')
````

````python
# 다시 숫자 Index로 되돌리기
# Reset 하기
bank_df.reset_index
bank_df
````

````python
# First Name 열을 DataFrame의 Index로 생성하고 싶을때, csv 파일을 import하면서 한번에 수행이 가능.
# index_col= 파라미터를 정의한 다음, 해당 파라미터의 값으로 Index로 할 열을 지정하면 됨.
bank_df_two = pd.read_csv('bank_client_information.csv',index_col='First Name')
bank_df_two
````

````python
bank_df_three = pd.read_csv('bank_client_information.csv',index_col='Last Name')
bank_df_three
````

# DataFrame에서 열 선택하기
DataFrame에서 특정 열을 선택해서 계산에 활용하는 방법에 대해 기술합니다.
- ***DataFrame에서 특정열만 뽑아오면, Pandas Series가 됨.***

아래처럼 간단하게 참조 메서드로 (.) DataFrame 에서 특정 열을 가져올 수 있음

그러나 해당 방법은, 열 이름에 공백이있다면 작동하지 않기 때문에 사용하지 않음

- ```bank_df.Email```

따라서 대괄호를 사용해 꺼내오는것이 정석적 방법

```
sample = bank_df['Email']
sample
```

### 2개 이상의 열을 꺼내오는 방법은
대괄호를 두번 작성해야 함. 리스트로 값을 꺼내야하기 떄문
- 이것은 2차원 이상이기 때문에. DataFrame
```
sample = bank_df[['Email', 'First Name']]
sample
```

````python
# test용 DataFrame Import
bank_df = pd.read_csv('bank_client_information.csv')
bank_df
````

````python
# Email 열만 꺼내오기
# DataFrame에서 특정열만 뽑아오면, Pandas Series가 됨.
sample = bank_df['Email']
sample
````

````python
# 꺼내온 것은 Series 로 확인할 수 있음.
print(type(sample))
````

````python
# 아래처럼 간단하게 참조 메서드로 (.) DataFrame 에서 특정 열을 가져올 수 있음
# 그러나 해당 방법은, 열 이름에 공백이있다면 작동하지 않기 때문에 사용하지 않음
bank_df.Email
````

````python
# 2개 이상의 열을 꺼내오는 방법
# 대괄호를 두번 작성해야 함. 리스트로 값을 꺼내야하기 떄문
sample = bank_df[['Email', 'First Name']]
sample
````

````python
# Type을 확인해보면, DataFrame인것을 확인할 수 있음.
print(type(sample))
````

````python
# 변수로 미리 꺼낼 열 리스트를 할당해놓고, 파라미터에 넣는것 또한 가능함.
my_selected_column = ['First Name', 'Net Worth']

sample = bank_df[my_selected_column]
sample
````

````python
# DataFrame에서 특정 행에 접근할 때, 슬라이스를 활용할 수 있음.
# 0번째부터 2개 출력
bank_df[0:2]
````

````python
# 역순으로 출력
bank_df[::-1]
````
