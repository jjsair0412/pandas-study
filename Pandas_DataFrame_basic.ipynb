# Broadcasting 연산과 새로운 데이터프레임 값 설정 방법
Brodcasting 연산 방법과 새로운 데이터 프레임 값을 설정하는 방법에 대해 기술합니다.

Brodcasting 연산이란, **특정 칼럼(열) 에 있는 모든값에 대해서 영향을 주게끔하는 방법** 을 이야기 합니다.
- 예를들어 모든 고객의 순 자산이 동등하게 증가햇을 경우, 모든 고객에 대해 같은 값을 늘려주기를 브로드케스팅 연산으로 할 수 있습니다.


````python
# 테스트용 DataFrame Import
bank_df = pd.read_csv('bank_client_information.csv')
bank_df
````

````python
# 특정 칼럼(열) 에 있는 모든값에 대해서 영향을 주게끔하는 방법
# 아래처럼 사용하면, 'Net Worth' 컬럼에 대해 모두 1000 이 증가함
bank_df['Net Worth'] = bank_df['Net Worth'] + 1000
bank_df
````

````python
# .add() 메서드 또한 사용할 수 있음.
# bank_df['Net Worth'] = bank_df['Net Worth'] + 1000 와 같은 결과값
bank_df['Net Worth'] = bank_df['Net Worth'].add(1000)
bank_df
````

````python
# .mul() 함수와 새로운 컬럼을 만드는것을 같이 수행할 수 도 있음.
# 'Net Worth (CAD)' 컬럼을 'Net Worth' 컬럼값으로 복사해서 생성하는데, 복사할 때 모든 값에 1.3을 곱해서 복사함
bank_df['Net Worth (CAD)'] = bank_df['Net Worth'].mul(1.3)
bank_df
````

# Pandasd 연산과 DataFrame 필터링
DataFrame의 데이터에 필터를 적용하는 방법에 대해 기술함.

조건문 (if) 구문처럼 , 특정 조건을 만족하는 컬럼만 꺼낼 수 있음.

```
# 주어진 조건을 만족하는 컬럼 선택
# Years with Bank 값이 10 이상 인 컬럼만 선택하는 방법
# {DataFrame_Name}[(조건문)] 형태로 사용하면 됨.
df_loyal = bank_df[ (bank_df['Years with Bank'] >= 10) ]
df_loyal
```

조건문이 여러개가 필요하다면, 논리연산자(AND , OR , ==) 을 통해 값을 꺼내면 됨.
```
# 조건문이 여러개일 경우 미리 조건문을 변수에 담아두고 논리연산자로 비교하면 됨.
# AND : & , OR : | , Equal : ==
mask_1 = bank_df['Years with Bank'] >= 10
mask_2 = bank_df['Net Worth'] >= 50000

df_loyal = bank_df[ mask_1 & mask_2 ]
df_loyal
```

논리연산자 없이, OR 연산을 .isin() 메서드로 할 수 있음.
```
# 전체 DataFrame 에서 Last Name이 Steve 또는 Mo 인 행을 꺼내옴.
# .isin() 메서드로 가능
mask = bank_df['Last Name'].isin(['Steve', 'Mo'])
bank_df[mask]
```

범위를 지정해서 필터링할 수 있음. .between() 메서드로 함.
```
# Net Worth가 5000부터 9000 사이에 있는 행을 선택하기
bank_df[bank_df['Net Worth'].between(5000, 9000)]
```

중복값들을 선택해서 출력할 수 있음. .duplicated() 메서드 사용
```
# 중복값이 있는 행을 선택하기
duplicet_result = bank_df['Last Name'].duplicated(keep = False)
bank_df[duplicet_result]
```

~ 연산자를 필터 앞에 두어서, 뒤에값의 True를 False로 둘 수 있음
```
# 중복값이 없는 값을 True로 두어서, 중복값이 아닌 값들만 출력
# 중복값이 있는 행을 선택하기
duplicet_result = ~bank_df['Last Name'].duplicated(keep = False)
bank_df[duplicet_result]
```

.drop_duplicates() 메서드를 통해 특정 컬럼에 중복값이 있다면, 제거하고 하나만 return
```
# .drop_duplicates() 메서드를 통해 선택한 특정 컬럼의 중복값이 있다면 , 중복값중 하나를 삭제하고 리턴함.

bank_df

bank_df.drop_duplicates(subset = ['Last Name'], inplace= True)
```

# Feature Engineering과 누락된 DataSet 처리 방안
실제 데이터들은 자료가 없거나 정해진 순서가 없고, 마음대로 흩어져있는 경우가 대다수임.

따라서 흩어져있거나 누락된 정보가 있는 DataSet들을 정재하는것이 중요하고, 어려움.

누락되거나 흩어져있는 정보들을 가진 DataSet을 정재하는 방안을 **Feature Engineering** 이라 하며, 이 방법을 한단계씩 따라하며 어떤식으로 처리하는지 기술함.

아래 기술된 관리방법으로 두가지가 있음.

- 1. 결측치(null인 값) 이 있는 행 자체를 제거한는 방법
- 2. 결측치(null인 값) 이 있는 행x열 위치를 찾아서 , 특정 데이터로 채워주는 방법

````python
# 첫번쨰로, null 요소가 있는지 확인
employee_df.isnull()
````

````python
# .isnull() 결과로 반환받은 모든 값의 합
# False : 0 , True : 1
# 결괏값 확인 결과 특정 컬럼에 몇개나 null이 들어가 있는 지 확인할 수 있음.
# 총 10개 컬럼에 null이 들어가있음.
employee_df.isnull().sum()
````

**category** DataType은 DataFrame의 크기를 줄여주는 효과가 있음.

DataSet을 더 효율적이고 최적화 시킬 수 있음.
