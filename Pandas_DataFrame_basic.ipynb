# 정의한 DataFrame Export 하기

```.to_~()``` 메서드로 Export 할 수 있으며. html, csv, json 등 다양한 타입을 지원함.
- 파라미터로는 저장할 파일 이름을 지정함

# DataFarme에서 라벨 기반으로 요소를 선택하는 방법
```.LOC()``` 메서드를 통해 DataFrame에서 특정 요소를 선택하는 방법에 대해 기술합니다.




````python
# test용 csv import
# index를 Last Name으로 사용
bank_df = pd.read_csv('bank_client_information.csv', index_col = 'Last Name')
bank_df
````

````python
# DataFrame을 알파벳 순으로 정렬
# .sort_index() 메서드로 DataFrame의 Index를 기반으로 알파벳순 정렬합니다.
bank_df.sort_index(inplace=True)
bank_df
````

````python
# loc 메서드를 통해서 , DataFrame의 특정 요소를 선택할 수 있음.
# loc 는 행과 열을 필터링할 수 있는데, 이를 라벨 기반의 인덱싱이라 함.
# 만약 정수 인덱스 기준으로 행 열을 필터링한다면, iloc 메서드를 사용함.

# 또한 아래 예시처럼 필터링 이후에 모든 raw중 필터링 결과 하나만 출력된다면, Series가 생성됨.
# 특정 label (Small) 인 친구를 꺼내옴
bank_df.loc['Small']
````

````python
# 만약 아래처럼 필터링 결과 두개 이상이라면, Data Frame이 출력됨.
bank_df.loc['Steve']
````

````python
# : , colon 을 통해서 특정 인덱스부터 특정 인덱스까지 값을 뽑을 수 있습니다.
# : 앞뒤도 다 선택됨.
# Ahmed 부터 Patton 까지 뽑아라. , Python slice와 같은 맥락임. (어디서부터 어디까지 뽑아라.)
bank_df.loc['Ahmed':'Patton']
````

````python
# 처음부터 Keller 인덱스까지 다 갖고와라 라는의미.
# Slice와 동일
bank_df.loc[:'Keller']
````

````python
# 여러개의 행이 대상이 된다면, 리스트 타입으로 지정해주면 됨.
# Steve 열은 복수라 다 출력된거임.
bank_df.loc[['Keller', 'Steve', 'Mo']]
````

````python
# DataFrame 일부를 무작위로 선택할 수 잇음
# n = ? 값으로 몇개를 랜덤으로 갖고올지를 선택
# 여기에도 axis 값이 들어가는데, 0은 행을 기준으로 동작하고 1은 열을 기준으로 동작함.
# 0이면 행중 5개랜덤, 1이면 열중 5개 랜덤
bank_df.sample(n = 5, axis = 0)
````

````python
# DataFrame의 특정 퍼센트정도만 랜덤으로 가져오는것도 가능함.
# DataFrame의 30%만 꺼내오기
bank_df.sample(frac= 0.3, axis = 0)
````

# .iloc , 정수기반 인덱스 필터링

```.loc``` 메서드는 인덱스가 정수가 아닐경우에 사용함.

그런데 정수일 경우 인덱스를 필터링해주는 ```.iloc``` 메서드에 대해 기록해 두었음.

````python
# test csv import
bank_df = pd.read_csv('bank_client_information.csv')
bank_df
````

````python
# 정수형 index를 통해 특정 index를 추출함.
# .iloc 메서드를 사용함.
# 가장 단순하게 하나만 뽑고싶다면 .iloc[] 파라미터로 특정 index의 정숫값을 넣어주면 됨.
bank_df.iloc[9]
````

````python
# .iloc 또한 python slice 기법을 지원함.
# 2:5 로 2번부터 5번 직전까지 추출
# 직전까지라는 의미가 중요함.
# .loc 은 : 기준으로 앞뒤가 다 추출되는데, iloc 은 : 뒤에값 직전까지만 추출함.
bank_df.iloc[2:5]
````

````python
# 인덱스 4행까지의 모든 행 가져오기
# :4 로 0,1,2,3 까지 추출
bank_df.iloc[:4]
````

````python
# DataFrame에서 여러 요소 의 index 추출
# 이것또한 .loc 과 동일하게 파라미터로 리스트를 넣어주면 됨.
bank_df.iloc[[2,4,9]]
````

````python
# DataFrame에서 특정 열과 행을 지정해서 슬라이싱하는것 또한가능.
# 특정 행으로 가서 특정 열만 가져오고싶은것임.
# 아래는 4행으로 가서 , 0번째 열부터 3번째 열 직전, 0,1,2 열을 가져와라 라는 의미
# 이는 결괏값이 한개만 있기 때문에 Pandas Series로 출력되는것을 확인할 수 있음.
bank_df.iloc[4,0:3]
````

````python
bank_df.iloc[[8,9]]
````

````python
bank_df.iloc[:7:-1]
````

````python
bank_df.iloc[-2:]
````
